```python
import polars as pl

df_child = pl.read_excel("data/policy_tracker.xlsx", sheet_name="4. US State - Current Child")

print(f"Number of rows: {df_child.shape[0]}")
print(f"Number of columns: {df_child.shape[1]}")
print(f"\nColumn names:")
print(df_child.columns)
```

```python
df_child.head()
```

```python
bills_by_state = (
    df_child
    .group_by("US State")
    .agg(pl.count().alias("num_bills"))
    .sort("num_bills", descending=True)
)

print(bills_by_state)
```

```python
total_bills = len(df_child)
total_states = df_child.select("US State").n_unique()

print(f"Total child safety bills: {total_bills}")
print(f"Number of states with bills: {total_states}")
```

```python
status_counts = (
    df_child
    .group_by("Status (optional)")
    .agg(pl.count().alias("num_bills"))
    .sort("num_bills", descending=True)
)

print(status_counts)
```

```python
# Look at the three issue tag columns
issue_tag_counts = (
    df_child
    .group_by("Issue Tag")
    .agg(pl.count().alias("num_bills"))
    .sort("num_bills", descending=True)
)

print("=== Primary Issue Tag ===")
print(issue_tag_counts)
```

```python
print("=== Issue Tag (primary) ===")
print(df_child.group_by("Issue Tag").agg(pl.count()).sort("count", descending=True).head(10))

print("\n=== Issue Tag_1 (secondary) ===")
print(df_child.group_by("Issue Tag_1").agg(pl.count()).sort("count", descending=True).head(10))

print("\n=== Issue Tag_2 (tertiary) ===")
print(df_child.group_by("Issue Tag_2").agg(pl.count()).sort("count", descending=True).head(10))
```

```python
# Show 5 random bills with their names and descriptions
sample = df_child.select(["US State", "Name", "Description"]).sample(5)

for row in sample.iter_rows(named=True):
    print("="*80)
    print(f"State: {row['US State']}")
    print(f"Name: {row['Name']}")
    print(f"Description: {row['Description']}")
    print()
```

```python
# Count bills that mention specific policy mechanisms
age_verify = df_child.filter(pl.col("Description").str.contains("(?i)age verification"))
parental_consent = df_child.filter(pl.col("Description").str.contains("(?i)parental consent|parent.*consent"))
data_privacy = df_child.filter(pl.col("Description").str.contains("(?i)data|personal information|privacy"))
content_filter = df_child.filter(pl.col("Description").str.contains("(?i)filter|block|restrict content"))
school_based = df_child.filter(pl.col("Description").str.contains("(?i)school|education|student"))

print(f"Age verification: {len(age_verify)} bills")
print(f"Parental consent: {len(parental_consent)} bills")
print(f"Data/privacy: {len(data_privacy)} bills")
print(f"Content filtering: {len(content_filter)} bills")
print(f"School-based: {len(school_based)} bills")
```

```python
# Create a summary showing which policy approaches each state is using
state_approaches = df_child.select([
    "US State",
    pl.col("Description").str.contains("(?i)age verification").alias("age_verify"),
    pl.col("Description").str.contains("(?i)parental consent|parent.*consent").alias("parental_consent"),
    pl.col("Description").str.contains("(?i)data|personal information|privacy").alias("data_privacy"),
    pl.col("Description").str.contains("(?i)filter|block|restrict content").alias("content_filter"),
]).group_by("US State").agg([
    pl.col("age_verify").sum().alias("age_verification_bills"),
    pl.col("parental_consent").sum().alias("parental_consent_bills"),
    pl.col("data_privacy").sum().alias("data_privacy_bills"),
    pl.col("content_filter").sum().alias("content_filter_bills"),
]).sort("data_privacy_bills", descending=True)

print(state_approaches.head(15))
```

```python
# Get bills that mention data/privacy and show their descriptions
data_privacy_bills = df_child.filter(
    pl.col("Description").str.contains("(?i)data|personal information|privacy")
).select(["US State", "Name", "Description"])

# Show 5 examples
print("=== EXAMPLES OF DATA PRIVACY BILLS ===\n")
for i, row in enumerate(data_privacy_bills.head(5).iter_rows(named=True)):
    print(f"{i+1}. {row['US State']}: {row['Name']}")
    print(f"   {row['Description']}\n")
```
